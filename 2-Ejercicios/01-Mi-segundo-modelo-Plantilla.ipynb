## Describir Escaladores

### Subtask:
Explicar el funcionamiento de los escaladores `MaxAbsScaler`, `RobustScaler`, `Normalizer`, `QuantileTransformer` y `PowerTransformer`.


### MaxAbsScaler

El `MaxAbsScaler` escala cada característica de forma independiente dividiendo cada valor por el valor absoluto máximo de esa característica. Esto da como resultado que los datos escalados estén dentro del rango `[-1, 1]`. Es útil cuando los datos ya están centrados en cero o contienen valores dispersos, ya que no altera la dispersión y evita distorsionar la distribución original, a diferencia de `MinMaxScaler` que sí afecta la dispersión al ajustar los datos en un rango fijo `[0, 1]` o `[-1, 1]` según sea el caso. Es particularmente adecuado para datos donde se sabe que la media es cero y se desea preservar la información sobre la escasez, como en el procesamiento de datos dispersos.

### RobustScaler

El `RobustScaler` elimina la mediana y escala los datos seg\u00fan el rango intercuart\u00edlico (IQR). El IQR es la diferencia entre el percentil 75 (Q3) y el percentil 25 (Q1). Este escalador es \u00fatil en la presencia de _outliers_ (valores at\u00edpicos), ya que utiliza estad\u00edsticos robustos que son menos sensibles a ellos. A diferencia de `StandardScaler`, que utiliza la media y la desviaci\u00f3n est\u00e1ndar (ambos fuertemente influenciados por los _outliers_), `RobustScaler` garantiza que los valores at\u00edpicos tengan un impacto m\u00ednimo en la transformaci\u00f3n. Los datos se escalan para que la mediana de cada caracter\u00edstica sea cero y el IQR sea uno, lo que ayuda a prevenir que los _outliers_ dominen el rango de los datos.

### Normalizer

El `Normalizer` escala cada muestra (fila) de datos de forma individual para que tenga una norma unitaria. Esto significa que la suma de los cuadrados de los valores de cada fila (si se usa la norma L2) ser&#225; 1. A diferencia de otros escaladores que operan sobre las caracter&#237;sticas (columnas), `Normalizer` act&#250;a sobre las muestras, lo que es &#250;til cuando se trabaja con datos donde la direcci&#243;n de los vectores de caracter&#237;sticas es m&#225;s importante que su magnitud absoluta. Se usa com&#250;nmente en algoritmos como el clustering k-means, la miner&#237;a de texto (tf-idf) o en redes neuronales donde la entrada por muestra debe tener una norma espec&#237;fica. Este escalador no elimina la media ni cambia la forma de la distribuci&#243;n, solo ajusta la magnitud de los vectores.

### QuantileTransformer

El `QuantileTransformer` transforma las caracter&#237;sticas para que sigan una distribuci&#243;n espec&#237;fica (uniforme o normal/Gaussiana). Utiliza el rango de cuantiles para mapear los valores de los datos a una nueva distribuci&#243;n. Primero, estima una funci&#243;n de transformaci&#243;n no lineal basada en los cuantiles de los datos. Luego, mapea cada punto de dato a un valor en la distribuci&#243;n de salida (por defecto, una distribuci&#243;n uniforme en el rango [0, 1]). Alternativamente, se puede especificar una distribuci&#243;n normal (`output_distribution='normal'`) para que los datos transformados sigan una distribuci&#243;n Gaussiana. Esta transformaci&#243;n es &#250;til para mitigar el impacto de los _outliers_ y para hacer que los datos sean m&#225;s adecuados para modelos que asumen una distribuci&#243;n normal, como muchos algoritmos lineales. Es una transformaci&#243;n robusta que puede manejar diferentes tipos de distribuciones de datos de entrada.

### PowerTransformer

El `PowerTransformer` es un tipo de escalador que transforma los datos para que sigan una distribuci&#243;n m&#225;s gaussiana (normal), lo que es beneficioso para muchos modelos predictivos que asumen normalidad en los datos de entrada. Esto lo logra aplicando una transformaci&#243;n de potencia, que puede ser la transformaci&#243;n de Box-Cox o la transformaci&#243;n de Yeo-Johnson. Ambas transformaciones son param&#233;tricas y buscan estabilizar la varianza y minimizar la asimetr&#237;a (skewness) en la distribuci&#243;n de los datos. La transformaci&#243;n de Box-Cox solo funciona con datos positivos, mientras que la de Yeo-Johnson puede aplicarse a datos que contengan valores cero o negativos. Esta t&#233;cnica es &#250;til para mejorar el rendimiento de modelos que son sensibles a la distribuci&#243;n de los datos, como la regresi&#243;n lineal, el an&#225;lisis discriminante lineal y los modelos basados en Gaussianas.

## Evaluar MaxAbsScaler

### Subtask:
Aplicar MaxAbsScaler a los conjuntos de entrenamiento y prueba, entrenar un modelo de regresión lineal y calcular el MSE.


**Reasoning**:
I need to apply the MaxAbsScaler to the training and test sets, train a Linear Regression model, and calculate the Mean Squared Error (MSE) to evaluate this scaling method.



```python
from sklearn.preprocessing import MaxAbsScaler

# 1. Inicializar MaxAbsScaler
maxabs_scaler = MaxAbsScaler()

# 2. Ajustar el escalador a X_train y transformar X_train y X_test
X_train_scaled_maxabs = maxabs_scaler.fit_transform(X_train)
X_test_scaled_maxabs = maxabs_scaler.transform(X_test)

# 3. Crear una nueva instancia de LinearRegression
model_maxabs = LinearRegression()

# 4. Entrenar el modelo con los datos escalados
model_maxabs.fit(X_train_scaled_maxabs, y_train)

# 5. Realizar predicciones sobre los datos de prueba escalados
y_pred_maxabs = model_maxabs.predict(X_test_scaled_maxabs)

# 6. Calcular el Error Cuadrático Medio (MSE)
mse_maxabs_scaler = mean_squared_error(y_test, y_pred_maxabs)

print(f"MSE with MaxAbsScaler: {mse_maxabs_scaler}")
```
